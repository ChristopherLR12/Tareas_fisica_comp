## Solución

I) Recordemos que la ecuación de una línea recta se puede escribir como:

$$y=mx+b$$

donde $m$ es la pendiente y $b$ la ordenada al origen, entonces obtenemos la pendiente

$$m=\frac{f(x_i)-f(x_{i-1})}{x_{i}-x_{i-1}}$$

Para obtener la ordenada al origen se sustituye lo anterior con el punto $(x_i, f(x_i))$ y se tiene que

$$f(x_i)=\frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}x_i+b$$
$$ \Rightarrow b=f(x_i)-\frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}x_i $$

por tanto

$$f(x)= \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}x+f(x_i)-\frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}x_i $$
$$ \Rightarrow f(x)= \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}(x-x_i) +f(x_i)$$

Después para obtener el área del trapecio se integra el resultado anterior

$$\int_{x_i-1}^{x_i} \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}(x-x_i)+f(x_i) \: dx = \int_{x_i-1}^{x_i} \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}x  \: dx - \int_{x_i-1}^{x_i} \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}x_i + \int_{x_i-1}^{x_i} f(x_i)  \: dx =$$

$$  \frac{f(x_{i})-f(x_{i-1})}{x_{i-1}-x_{i}} \int_{x_i-1}^{x_i}x  \: dx - \frac{f(x_{i})-f(x_{i-1})}{x_{i-1}-x_{i}}x_i \int_{x_i-1}^{xi}\: dx + f(x_i) \int_{x_i-1}^{x_i}  \: dx =$$

$$  \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}} \frac{1}{2}(x_i^2-x_{i-1}^2) - \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}x_i (x_i-x_{i-1}) + f(x_i)(x_{i}-x_{i-1}) =$$

$$  \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}} \left(\frac{1}{2}(x_i^2-x_{i-1}^2) - x_i^2+x_ix_{i-1}\right) + f(x_i)(x_{i}-x_{i-1}) =$$

$$  \frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}} \left(\frac{1}{2}(x_i^2-x_{i-1}^2- 2x_i^2+2x_ix_{i-1})\right) + f(x_i)(x_{i}-x_{i-1}) =$$

$$  -\frac{1}{2}\frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}} (x_{i}-x_{i-1})^2 + f(x_i)(x_{i}-x_{i-1}) = -\frac{1}{2}\left(f(x_{i})-f(x_{i-1})\right)(x_{i}-x_{i-1}) + f(x_i)(x_{i}-x_{i-1}) =$$

$$  -\frac{(x_{i}-x_{i-1})}{2}\left(f(x_{i})-f(x_{i-1}) - 2f(x_i)\right) = -\frac{(x_{i}-x_{i-1})}{2}\left(-f(x_{i-1})-f(x_i)\right) = (x_{i}-x_{i-1})\frac{f(x_{i-1})+f(x_i)}{2}$$

````python
#II)

def areaTrapecio(f,a,b):
    return (b-a)*(f(a)+f(b))*(1/2)

def intTrapecio(f,a,b,n):
    xx = np.linspace(a,b,n+1)
    integral = 0
    for i in range(n):
        integral = integral + areaTrapecio(f,xx[i],xx[i+1])
    return integral
````

## Solución

````python
#I)

def diferenciaAdelante(f,a,h):
    return (f(a+h) - f(a))/h

def diferenciaAtras(f,a,h):
    return (f(a) - f(a-h))/h

def diferenciaCentral(f,a,h):
    return (f(a+h)-f(a-h))/(2*h)

σ = 1
ϵ = 4

def f2(r):
    return 4*ϵ*((σ/r)**12 - (σ/r)**6)

xx = np.linspace(-0.9,2,150)

yy_ad = np.array([diferenciaAdelante(f2,a,0.001)  for a in xx])
yy_at = np.array([diferenciaAtras(f2,a,0.001)  for a in xx])
yy_ce = np.array([diferenciaCentral(f2,a,0.001)  for a in xx])

plt.figure(figsize=(10,8))
plt.plot(xx,yy_ad,label = "Derivada hacia adelante",lw=2)
plt.plot(xx,yy_at,label= "Derivada hacia atrás",lw=2)
plt.plot(xx,yy_ce,label= "Derivada centrada",lw=2)
plt.xlabel("$r$",fontsize=15)
plt.ylabel("$f\'(r)$",fontsize=15)
plt.grid()
plt.legend()
plt.title("Distintos tipos de derivadas de $f(r)$",fontsize=20)
plt.show()
````

````python
#II)

def fa2(r):
    return (-192+96*r**6)/(r**13)

yy_a = np.array([fa2(a) for a in xx])

plt.figure(figsize=(10,8))
plt.plot(xx,abs(yy_ad-yy_a),label = "Error adelante",lw=2)
plt.plot(xx,abs(yy_at-yy_a),label= "Error atrás",lw=2)
plt.plot(xx,abs(yy_ce-yy_a),label= "Error central",lw=2)
plt.xlabel("$r$",fontsize=15)
plt.ylabel("Error",fontsize=15)
plt.grid()
plt.legend()
plt.title("Error en derivadas",fontsize=20)
plt.show()
````

````python
#III)

hh = np.logspace(-4,-1,num=300)

e_ad=[]
e_at=[]
e_ce=[]

for h in hh:
    e_ad.append(np.mean(abs(np.array([diferenciaAdelante(f2,a,h)  for a in xx])-np.array([fa2(a) for a in xx]))))
    
for h in hh:
    e_at.append(np.mean(abs(np.array([diferenciaAtras(f2,a,h)  for a in xx])-np.array([fa2(a) for a in xx]))))
    
for h in hh:
    e_ce.append(np.mean(abs(np.array([diferenciaCentral(f2,a,h)  for a in xx])-np.array([fa2(a) for a in xx]))))
    
plt.figure(figsize=(10,8))
plt.plot(hh,e_ad,label="Error adelante",lw=2)
plt.plot(hh,e_at,label= "Error atrás",lw=2)
plt.plot(hh,e_ce,label="Error centrada",lw=2)
plt.xlabel("$h$ (log)",fontsize=15)
plt.ylabel("Promedio de errores (log)",fontsize=15)
plt.xscale("log")
plt.yscale("log")
plt.grid()
plt.legend()
plt.title("Error en derivadas respecto a $h$",fontsize=20)
plt.show()

plt.figure(figsize=(10,8))
plt.plot(hh,e_ad,label="Error adelante",lw=2)
plt.plot(hh,e_at,label= "Error atrás",lw=2)
plt.plot(hh,e_ce,label="Error centrada",lw=2)
plt.xlabel("$h$",fontsize=15)
plt.ylabel("Promedio de errores (log)",fontsize=15)
plt.yscale("log")
plt.grid()
plt.legend()
plt.title("Error en derivadas respecto a $h$",fontsize=20)
plt.show()
````

(i) Define una función `trayectoria_t` que calcula la trayectoria de una partícula en tiro parabólico en 2 dimensiones con una posición $(x_0, y_0)$ y rapidez $v_0$ iniciales dadas y lanzado a un ángulo $\theta$ sobre la horizontal. `pos_t` debe tomar como argumentos una lista y tres valores, el primero con la posición inicial (que es un vector, i.e. un arreglo de dos entradas en este caso), el segundo con la rapidez inicial,  el tercero el ángulo $\theta$ y el cuarto un tiempo hasta el que se calcula la trayectoria. Lo que tu función debe regresar son dos arreglos, el primero con las coordenadas horizontales ($x(t)$) y el segundo con las coordenadas verticales ($y(t)$) para cada punto de la trayectoria.

Utiliza las fórmulas analíticas conocidas para ello.

$$\begin{align}
x(t) &= x_0 + v_{x0}t\\
y(t) &= y_0 + v_{y0}t - \frac{1}{2}gt^2\\
\end{align}$$

**Sugerencia 1:** Tendrás que definir dentro de tu función un arreglo con los tiempos en los que calcularás las posiciones, esto puedes hacerlo usando `np.linspace(0,t,<número de tiempos a calcular>`. Ten cuidado con usar suficientes puntos para que la gráfica sea suave.

**Sugerencia 2:** Deberás usar las funciones `np.sin` y `np.cos` que calculan el seno y coseno de su argumento. Ojo, a estas funciones hay que pasarles el argumento en radianes, entonces debes decidir si le pasaras $\theta$ a tu función en grados o en radiantes y dependiendo de lo que decidas, quiza tendrás que hacer una conversión de unidades dentro de tu función.

````python
#I)

def trayectoria_t(lp,v0,θ,tc):
    g=9.81
    tt=np.linspace(0,tc,1000)
    xx=[lp[0]+v0*np.cos(θ)*t for t in tt]
    yy=[lp[1]+v0*np.sin(θ)*t-(1/2)*g*t**2 for t in tt]
    return xx,yy
````
