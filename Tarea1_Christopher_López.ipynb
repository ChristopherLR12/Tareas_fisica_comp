# Iteraciones: ciclos for y while

Una caracteristica importante de los algoritmos computacionales es la habilidad que nos proporcionan de llevar a cabo c치lculos de forma repetida que tomar칤an un muy largo tiempo de realizar a mano. Llamamos **iteraci칩n** al proceso de repetir ciertas instrucciones para distintos objetos. Esto se refiere a que cierta variable recorra un conjunto de valores, a trav칠s de un **bucle** (ciclo o _loop_).

Empezaremos discutiendo los ciclos `for` ("para", en espa침ol) y continuaremos con los ciclos `while` ("mientras", en espa침ol).

Ya hemos visto las comprensiones de arreglo, que proveen una forma sencilla y r치pida para crear arreglos, y ocupan la palabra `for`. Los bucles `for` nos permiten realizar iteraciones sobre un conjunto de variables previamente definidas, ya sea sobre un indice que etiqueta a los elementos de un conjunto o sobre los elementos mismos. Es decir, realizar un proceso **para** cada elemento de un conjunto.

Los ciclos `while` por su lado, nos permiten realizar iteraciones **mientras** se cumpla cierta condici칩n, es decir mientras cierta pregunta con respuesta s칤 o no pueda ser contestada y adem치s la respuesta sea s칤. En este sentido, los ciclos `while` no tienen una cantidad de pasos definida como los ciclos `for`, por lo que hay que tener cuidado con la manera en la que son programados para evitar un bucle infinito.

### Ciclos "for"

La sintaxis en `Python` para los ciclos `for` es la siguiente:

```
for <variable> in <arreglo>:
    [haz esto]
    [y esto]
```

En un bucle de este tipo, la variable toma cada valor en turno del arreglo. Se refiere a los comandos adentro como el "cuerpo" del bucle. Estos comandos se ejecutar치n una y otra vez, hasta que se agote el iterable.

Por lo tanto, este tipo de bucle se ocupa cuando se conoce de antemano el n칰mero de veces que se requiera repetir algo.

Aqu칤 un ejemplo:

````python
for i in range(-5,10):
    print(i**2)
print("ya me sal칤 del `for`")
````

Vemos que la variable `i` toma cada valor en el rango, de -5 hasta 9, uno por uno. Recordemos que `range(a,b)`, es una funci칩n de `python` que crea un arreglo iterable que va desde `a` hasta `b-1` de 1 en 1 y con `a`, `b` enteros. En un lenguaje "m치s matem치tico", `range(a,b)` representa $[a,b)\subset\mathbb{Z}$.

En el ejemplo anterior, se iter칩 sobre un objeto de typo `range`. Tambi칠n se puede iterar sobre una lista o una tupla:

````python
for num in [3, 4, 7]:
    print(num)
print("termin칠 el primer bucle")
for num in ("Crhis", "Leo", "Todxs lxs demxs"):
    print(num,"\n")
print("termin칠 el segundo bucle")
````

Recordemos que los n칰meros primos son los enteros positivos mayores que $1$ que s칩lo son divisibles exactamente entre $1$ y s칤 mismo.

````python
#I)

def esPrimo(n):
    mod_n=n%n
    mod_1=n%1
    
    l=[n%i for i in range(2,n-1)]
    
    if mod_n == 0 and mod_1 == 0:
        if 0 in l:
            return False
        else:
            return True
    else:
        return False
    
esPrimo(31)
````

````python
#II)

def nPrimos(n):
    primos=[]
    
    for i in range(2,n+1):
        if esPrimo(i) == True:
            primos.append(i)
        else:
            continue
    return(primos)

nPrimos(50)
````
