# Ecuaciones Diferenciales Ordinarias

Recordemos que una **ecuación diferencial ordinaria (EDO)** es una ecuación en la cual aparece una o más derivada de una función, y para la cual queremos buscar la función que satisface esta relación.

La EDO más simple no-trivial y físicamente relevante en una variable es 

$$\dot{x} = \alpha \, x. \qquad (*)$$

Recordemos la notación $\dot{x} := \textstyle \frac{dx}{dt}$, donde $t$ es el tiempo; es muy común en la física que las EDOs involucren derivadas con respecto al tiempo.

Una solución a esta ecuación es una función $x$, tal que cuando la derivamos con respecto al tiempo, nos arroja $\alpha$ (una constante) multiplicada por la misma función. Implícitamente está claro que $x$ depende de $t$, por lo que realmente $x$ se refiere a una función de $t$, es decir $x: \mathbb{R} \to \mathbb{R}$ es la función $t \mapsto x(t)$. 

Por lo tanto, la ecuación $\dot{x} = \alpha x$ se puede considerar como una ecuación *funcional*, es decir, una igualdad entre funciones. 

Para entenderla, se reescribe de forma más explícita como sigue:

$$\dot{x}(t) = \alpha x(t) \qquad \text{para cada } t \in \mathbb{R}.$$

Ahora es una igualdad uno a uno de vores evaluados de $x(t)$ y $t$, que dice que la "velocidad" en el tiempo $t$ es un cierto múltiplo de la posición en ese momento.

Esto quiere decir que si *de alguna manera* hayamos logrado saber que la solución en el tiempo $t$ está en la posición $x(t)$, entonces nos indica cuál será la *velocidad* $\dot{x}(t)$ en este momento.

La forma general de una EDO en una variable es

$$\dot{x} = f(x, t),$$

o sea

$$\dot{x}(t) = f(x(t), t) \qquad \text{para cada } t \in \mathbb{R}.$$


### Problemas de valores iniciales

Recordemos que de ir acompañada por una condición inicial $x(t=t_0) = x_0$ (**problema de valores iniciales**, o **problema de Cauchy**), lo que obtenemos es una única solución (si existe).  

La solución de una ecuación de este tipo será una **función** $x(t)$ que satisface al mismo tiempo.

$$
\begin{cases}
\dot{x}(t) = f(x(t),t), \quad t \in I \\
x(t_0) = x_0, \\
\end{cases}
$$

con $I$ el dominio de la función.

En el caso particular de la ecuación $\dot{x} = -\alpha x$, conocemos analíticamente la solución, y nos servirá para comprobar nuestros métodos. Pero para las ecuaciones de interés para la física, esto casi nunca ocurre. En estos casos, debemos aplicar distintas técnicas de *aproximación* de la solución. Para ello usaremos *métodos numéricos* para encontrar valores de la solución en puntos específicos del dominio de la función.  
<br>


### Retrato fase

Antes de resolver una ecuación diferencial, podemos darnos una idea del comportamiento que tendrán las soluciones si graficamos el campo de direcciones: el campo vectorial de las pendientes de las funciones que son solución a la ecuación diferencial pero no necesariamente a la condición inicial. Dicho campo corresponde al campo vectorial $P:\mathbb{R}^2 \to \mathbb{R}^2$ definido por

$$
P(t,x) = (1,f(x,t))
$$

Para graficar este campo vectorial podemos usar las funciones `plt.quiver` y `plt.contour` de `matplotlib.pyplot`.

### Ejemplo 1:

Sea $\dot{x}(t) = \alpha x(t) $. De acuerdo a la notación usada hasta ahora, $f(x(t),t) = \alpha x(t) $. El campo vectorial entonces es

$$
P(t,x) = ( 1,\alpha x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

tt = np.linspace(-1,1,15)
xx = np.linspace(-2,2,15)
alpha = .5
vv_x = np.array([[alpha*x for t in tt] for x in xx])
vv_t = np.ones_like(vv_x)
plt.quiver(tt,xx,vv_t,vv_x)
plt.xlabel("$t$")
plt.ylabel("$x$")
```

# Método de Euler

Para resolver una EDO numéricamente en la computadora, tendremos que *aproximar* la solución continua $x: t \mapsto x(t)$. La manera más sencilla de proceder es utilizando [**diferencias finitas**](https://en.wikipedia.org/wiki/Finite_difference). Empezaremos con el método más sencillo llamado [método de Euler](https://en.wikipedia.org/wiki/Euler_method). Para esto, calcularemos la solución de la ecuación diferencial en un número finito de puntos del dominio, digamos,

$$\{t_0,t_1,...,t_n\}.$$
A esto le llamamos discretización del espacio. Si espaciamos uniformemente las $t_i$ en lo que llamaremos un paso de tiempo $h$, entonces tenemos que 

$$
\begin{align*}
t_0 & = t_0,\\
t_1 & = t_0 + h,\\
t_2 & = t_1 +h & = t_0 + 2h,\\
&... &...\\
t_n &= t_{n-1} + h & = t_0 + nh,
\end{align*}
$$

Y buscamos entonces los valores de $x(t_i)$ para cada $t_i$. Etiquetemos a estos valores como $x_i = x(t_i)$. De esta forma, tenemos una discretización de la función $x(t)$. Esto no es otra cosa más que tener un conjunto de evaluaciones de la solución $x(t)$.

#### Ejercicio 1

i) ¿Cuál es la aproximación más sencilla de la derivada $\dot{x}(t)$, en términos de un tamaño de paso pequeño $h$?

ii) Sustituye esta aproximación en la ecuación $\dot{x} = f(x, t)$ para obtener una expresión para $x_{i+1} = x(t_{i+1})$, es decir el valor aproximado en el siguiente paso de tiempo, en términos del valor ya conocido $x_i = x(t_i)$.

iii) Define una función `mi_euler` que tome como argumentos una función `f`, una condición inicial `x0`, un tiempo final `tf` y un paso de tiempo `h` (este que use un valor _default_ de 1e-4) y que regrese dos arreglos `tt`, `xx` donde `tt` es el intervalo de tiempo que va desde $t_0 = 0$ hasta $t_f$ en intervalos uniformes de $h$ mientras que `xx` son los $x_i$ solución de la ecuación diferencial 

$$\dot x(t) = f(x,t)$$ de acuerdo a lo visto hasta ahora.

```python
def mi_euler(f,x0,tf,h=1e-4):
    n = int(np.ceil(tf/h))
    tt = [i*h for i in range(n+1)]
    xx = [x0]
    
    for i in range(1,n+1):
        x_i = xx[i-1] + f(xx[i-1], tt[i-1]) * h
        xx.append(x_i)
    
    return tt,xx
```

```python
def f1(x,t):
    return alpha*x
```

```python
tt_sol,xx_sol = mi_euler(f1,1,5,h=.1)
```

## Ecuaciones de orden mayor: reducción

En la práctica del día a día, en particular en la física, por lo general tratamos con ecuaciones diferenciales de orden mayor a uno. En general, una ecuación diferencial de orden $l$ es una ecuación funcional de la forma

$$
y^{(l)} (t) = g(y^{(l-1)} (t),y^{(l-2)} (t), \ldots, y^{(1)}(t), y(t),t)
$$

Donde $y^{(l)} (t) = \frac{d^l y}{dt^l}$ es la $l$-ésima derivada de $y(t)$. Aunque podríamos desarrollar un método para atacar ecuaciones diferenciales de ese orden, existe un [**truco**](http://mathonline.wikidot.com/converting-nth-order-odes-to-systems-of-n-first-order-odes) para convertir una ecuación de orden $n$ en un sistema de ecuaciones de primer orden con $n$ variables. Definimos

$$
\begin{align}
 x_1(t) &= y(t) \\
 x_2(t) &= y^{(1)}(t) \\
 &\vdots \\
 x_i(t) &= y^{(i-1)}(t) \\
 &\vdots \\
 x_l(t) &= y^{(l-1)}(t).
\end{align}
$$

Entonces la ecuación diferencial se reduce al siguiente sistema

$$
\begin{align}
 \dot{x_1}(t) &= y^{(1)}(t) = x_2(t) \\
 \dot{x_2}(t) &= y^{(2)}(t) = x_3(t) \\
 &\vdots \\
 \dot{x_i}(t) &= y^{(i)}(t) = x_{i+1}(t) \\
 &\vdots \\
 \dot{x}_{l-1}(t) &= y^{(l-1)}(t) = x_{l}(t) \\
\dot{x_l}(t) &= y^{(l)}(t).
\end{align}
$$

Pero

$$
y^{(l)}(t) = g(y^{(l-1)} (t),y^{(l-2)} (t), \ldots, y^{(1)}(t),y(t),t) = g(x_{l}(t),x_{l-1}(t),\ldots,x_1(t),t).
$$

Por lo que el sistema se vuelve

$$
\begin{align}
 \dot{x_1}(t) &= x_2(t) \\
 \dot{x_2}(t) &= x_3(t) \\
 &\vdots \\
 \dot{x_i}(t) &= x_{i+1}(t) \\
 &\vdots \\
 \dot{x}_{l-1}(t) &= x_{l}(t) \\
\dot{x_l}(t) &= g(x_{l}(t),\ldots,x_1(t),t),
\end{align}
$$

o en notación vectorial

$$\dot{\mathbf{x}}(t) = G(\mathbf{x}(t),t),$$

con 

$$\mathbf{x}(t) = (x_1,x_2,...,x_l) = (y^{(1)},y^{(2)},...,y^{(l)}),$$
y
$$G(\mathbf{x}(t),t) = (\dot{x_1},\dot{x_2},...,\dot{x_l}) = (x_2,x_3,...,g(\mathbf{x},t)) = (y^{(2)},y^{(3)},...,g(\mathbf{x},t)).$$

## Método de euler vectorial

Sea entonces **el vector de funciones**  $\mathbf{x}(t) : \mathbb{R} \to \mathbb{R}^{l}$

$$\mathbf{x}(t) := (x_1(t),\ldots,x_{l}(t))$$ 

y la función vectorial $\mathbf{f}(\mathbf{x}(t),t) : \mathbb{R}^{l+1} \to \mathbb{R}^{l}$

$$\mathbf{f}(\mathbf{x}(t),t) = (x_2(t),\ldots,x_{l}(t),g(x_{l}(t),x_{l-1}(t),\ldots,x_1(t),t))$$ 

Podemos escribir el sistema de ecuaciones diferenciales de forma vectorial como

$$
\dot{\mathbf{x}}(t) = \mathbf{f}(\mathbf{x}(t),t)
$$

Que nuevamente admite una solución mediante el método de euler al aproximar la derivada del vector $\mathbf{x}(t)$ con una diferencia hacia adelante:

$$
\mathbf{x}(t+h) = \mathbf{x}(t) + h \; \mathbf{f}(\mathbf{x}(t),t)
$$

#### Ejercicio 6

Implementa una función `eulerVec(f,x_0,t_0,t_f,h)` con `f` la función **vectorial** tal que $\dot{\mathbf{x}}(t) = \mathbf{f}(\mathbf{x}(t),t)$, `x_0` (un vector de $n$ entradas) y `t_0` la condición inicial, `t_f` el tiempo máximo al que queremos obtener la solución y `h` el tamaño de paso. La función debe de implementar el método de euler y debe de regresar el arreglo de los tiempos `[t_0,t_1, t_2,....,t_n]` con $t_{n} \leq t_f < t_{n} + h$ y el de la solución a la ecuación diferencial `[xx_0,xx_1,...,xx_n]` (Nota que esto es una matriz).

```python
def eulerVec(f,xx0,t0,tf,h=1e-4):
    n = int(np.ceil((tf-t0)/h))
    tt = [i*h for i in range(n+1)]
    xxx = np.array([xx0])
    
    for ti in range(1,n+1):
        xx_i = xxx[ti-1] + f(xxx[ti-1],tt[ti-1]) * h
        xxx = np.append(xxx,[xx_i],axis=0)
    
    return np.array(tt),xxx
```
