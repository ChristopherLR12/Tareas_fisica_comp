# Solución de ecuaciones no lineales: búsqueda de raíces y optimización

_______

## Búsqueda de raíces

Muchas veces nos encontramos con el problema de encontrar el valor $x_*$ tal que cierta función $f: \mathbb{R} \to \mathbb{R}$ cumple que $f(x_*)=0$

## Ejemplos:

1. $x^2 -6x + 9 = 0$ $ \quad \quad \quad f(x) = x^2 -6x + 9$


2. $\sin{x} = 0$ $ \quad \quad \quad f(x) = \sin{x}$


3. $ x^2 - \sqrt[3]{\cos{7x}} = \tanh^{-1}{\left( e^{-x^2+1} \right)} \implies  x^2 - \sqrt[3]{\cos{7x}} - \tanh^{-1}{\left( e^{-x^2+1} \right)} = 0$ $ \quad \quad \quad f(x) = x^2 - \sqrt[3]{\cos{7x}} - \tanh^{-1}{\left( e^{-x^2+1} \right)}$

En ocaciones, podremos simplemente hacer un poco de álgebra para encontrar $x_*$, pero en otras esto es imposible (o quizá muy difícil). Si no existe una solución analítica, una de las cosas que podemos hacer es resolver el problema de manera **numérica**.

### Algoritmo (o Método) de Newton-Raphson

Supongamos que $f$ es diferenciable, es decir, su primera derivada $f'(x)$ existe y es continua. Podemos aproximar $f$ en una serie de Taylor a primer orden para un punto $x_*$, donde $f(x_*)=0$, entonces escribiendo $x=x_*+h$, con $h\ll 1$,

$$
f(x_*) = 0 \approx f(x) + f'(x)h.
$$

Supongamos que conocemos $x$ y lo que estamos buscando es cuál es el valor de $h$ tal que $x+h$ es una raiz de $f$. Usando la ecuación anterior, podemos expresar $h$

$$
h = -\frac{f(x)}{f'(x)}
$$

Esto nos permite proponer un esquema **recursivo** para construir una sucesión $\{x_i\}$ tal que converga a una solución de la ecuación.

$$
x_i = \begin{cases}
    x_{\mathrm{inicial}} & \text{ si } i=0 \\
    x_{i-1} + h = x_{i-1} -\frac{f(x_{i-1})}{f'(x_{i-1})} & \text{ si } i>0 \\
\end{cases}
$$

Construir dicha sucesión es el llamado **Método de Newton-Raphson**

```python
def newton_raphson(f,df,x_inicial,n):
    xx_nr = [x_inicial]
    for i in range(1,n):
        xx_nr.append(xx_nr[i-1] - f(xx_nr[i-1])/df(xx_nr[i-1]))
    return xx_nr
```

#### Ejemplo sencillo:
$f(x) = x^2 - 7 x + 12$  
$\frac{\mathrm{d}f(x)}{\mathrm{d}x} = f'(x) = 2x-7$

Analíticamente, sabemos que $f(x) = (x-3)(x-4)$, $\,f(3)=0\,$ y $\,f(4)=0$. Obtengamos el mismo resultado numéricamente:

```python
def f1(x):
    return x**2 - 7*x + 12
def df1(x):
    return 2*x - 7
```

```python
nr1 = newton_raphson(f1,df1,1,n=1000)
```

```python
plt.figure(figsize=(15,6))
plt.subplot(121)
plt.plot(nr1,[f1(x) for x in nr1],marker=".",label="{$x_k$}")
plt.xlabel("$x_i$",fontsize=15)
plt.ylabel("$f(x_i)$",fontsize=15)
plt.legend()
plt.grid()
plt.subplot(122)
plt.plot(nr1,[f1(x) for x in nr1],marker=".",label="{$x_k$}")
plt.xlabel("$x_i$",fontsize=15)
plt.ylabel("$f(x_i)$",fontsize=15)
plt.legend()
plt.yscale("log")
# plt.xscale("log")
plt.grid()
plt.show()
```

### ¿Qué es lo que hace (gráficamente)?

```python
xx = np.linspace(-1,8,100)
yy = [f1(x) for x in xx]
xxg1 = newton_raphson(f1,df1,3.55,n=7)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title("Algoritmo de Newton-Raphson")
ax.plot(xx,yy,label="$f1(x)$")
ax.grid()
ax.legend()
time.sleep(1)
plt.close()
for i in range(len(xxg1)-1):
    xc = xxg1[i]
    xcn = xxg1[i+1]
    fig, ax = plt.subplots(figsize=(10,5))
    ax.set_title("Algoritmo de Newton-Raphson")
    ax.plot(xx,yy,label="$f1(x)$")
    ax.grid()
    ax.legend()
    ax.scatter(xxg1[i],f1(xxg1[i]))
    plt.xlim(-1,8)
    plt.ylim(-.5,7)
    clear_output(wait=True)
    display(fig)
    time.sleep(1)
    plt.close()
    fig, ax = plt.subplots(figsize=(10,5))
    ax.set_title("Algoritmo de Newton-Raphson")
    ax.plot(xx,yy,label="$f1(x)$")
    ax.grid()
    ax.legend()
    ax.scatter(xc,f1(xc))
    plt.xlim(-1,8)
    plt.ylim(-.5,7)
    clear_output(wait=True)
    plt.plot([xc-6,xc,xc+6],[f1(xc)-df1(xc)*6,f1(xc),f1(xc)+df1(xc)*6],c="k",alpha=.5)
    plt.plot([xcn,xcn],[0,f1(xcn)],ls="--",c="k",alpha=.5,lw=2)
    display(fig)
    time.sleep(.8)
    plt.close()
plt.close()
```

## Optimización

Encontrar máximos y mínimos locales es otra tarea que puede ser de nuestro interés. Recordemos que los máximos y mínimos locales, en un intervalo $\mathcal{I}$ se definen como los puntos $\left(x_m,f\left(x_m\right)\right)$ y $\left(x_M,f\left(x_M\right)\right)$ tales que 

$$f(x_m)\leq f(x), \quad f(x_M)\geq f(x),\quad \forall x \in \mathcal{I},$$

para alguna función $f$ definida en el intervalo $\mathcal{I}$.

#### ¿Todas las funciones tienen mínimos/máximos locales?

Sí (_ish_). Dependerá del intervalo que estemos utilizando. Notemos que si usamos un intervalo cerrado, entonces, de no haber un máximo/mínimo en el interior del intervalo, alguno de los extremos cumplirá las condiciones mencionadas antes. Por otro lado, existe la posibilidad de tener un intervalo abierto (o semi abierto) de tal suerte que no exista un mínimo o máximo local.

#### ¿Es fácil decidir si una función tiene mínimos locales en su interior?

NO, tampoco es fácil decidir. Si $f$ es diferenciable y existen puntos en los que $\frac{\mathrm{d}f}{\mathrm{d}x} = 0$, pueden existir, pero no es seguro.

#### ¿Si existen, cómo encontrarlos?
  
Analíticamente no hay una manera algorítmica de hacerlo y depende del caso partícular lo que hay que hacer para hallarlos. Numéricamente, el problema es muy similar al de encontrar las raíces de una función.

#### Caso sencillo:

$f:\mathbb{R} \to \mathbb{R}$, es derivable y sabemos que tiene mínimos locales, por ejemplo,

$$f(x) = x^2 - 5x + 6$$
  
$$
f'(x) = 2x-5
$$

$f'(x)>0$ indica que $f$ es creciente

$f'(x)<0$ indica que $f$ es decreciente

```python
def f2(x):
    return x**2 - 5*x + 6
def df2(x):
    return 2*x-5
```

```python
xx = np.linspace(1,4)
yy = [f2(x) for x in xx]
dyy = [df2(x) for x in xx]
```

```python
plt.figure(figsize=(6,4))
plt.plot(xx,yy,label="$f(x) = x^2-5x+6$")
plt.plot(xx,dyy,label="$f\'(x) = 2x-5$",c="k",lw=1,alpha=.5)
plt.grid()
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.ylim(-2,2)
plt.legend()
plt.show()
```

La derivada nos indica hacia donde **crece** el valor de la función. Si yo quiero ir a un lugar donde **no crece**, es decir, donde es mínimo, me debo de mover en **dirección contraria a la derivada**.

Podemos usar esta idea para proponer un esquema recursivo para ir construyendo una sucesión que se acerque a un mínimo.

### Método del descenso de gradiente

Supongamos que tenemos un valor $x_{i}$ tal que está cerca de un mínimo. ¿Cómo nos acercamos más a ese mínimo?

$$
x_{i+1} = x_{i} - \alpha \cdot f'(x_{i})
$$

con $0 < \alpha < 1$. $\alpha$ es el tamaño de paso. Para poder poder implementar la sucesión, tengo que dar un punto inicial $x_{0}$ para crear la sucesión $\{x_i\}$ de acuerdo a 

$$
x_{i} = \begin{cases}
    x_0 & \text{si } i=0 \\
    x_{i-1} - \alpha \cdot f'(x_{i-1}) & \text{si } i>0 \\
\end{cases}
$$

```python
def desGrad(df,x0,alpha,n):
    xx_dg = [x0]
    for i in range(1,n):
        x_new = xx_dg[i-1] - alpha* df(xx_dg[i-1])
        xx_dg.append(x_new)
    return xx_dg
```

```python
desGrad(df2,3.0,0.1,20)
```

### ¿Qué es lo que hace (gráficamente)?

```python
xx = np.linspace(-5,10,200)
yy = [f2(x) for x in xx]
dyy = [df2(x) for x in xx]
alpha = .8
xxg2 = desGrad(df2,4.0,alpha,n=20)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title("Algoritmo del descenso de gradiente")
ax.plot(xx,yy,label="$f2(x)$")
ax.grid()
ax.legend()
time.sleep(1)
plt.close()
for i in range(len(xxg2)-1):
    xc = xxg2[i]
    xcn = xxg2[i+1]
    fig, ax = plt.subplots(figsize=(10,5))
    ax.set_title("Algoritmo del descenso de gradiente")
    ax.plot(xx,yy,label="$f_2(x_i)$")
    ax.plot(xx,dyy,label="$f_2'(x_i)$",c="k",lw=1,alpha=.5)
    ax.grid()
    ax.legend()
    ax.scatter(xxg2[i],f2(xxg2[i]))
    plt.xlim(.9,4.1)
    plt.ylim(-.5,2.1)
    clear_output(wait=True)
    display(fig)
    time.sleep(1)
    plt.close()
    fig, ax = plt.subplots(figsize=(10,5))
    ax.set_title("Algoritmo del descenso de gradiente")
    ax.plot(xx,yy,label="$f_2(x_i)$")
    ax.plot(xx,dyy,label="$f_2'(x_i)$",c="k",lw=1,alpha=.5)
    ax.grid()
    ax.legend()
    ax.scatter(xc,f2(xc))
    plt.xlim(.9,4.1)
    plt.ylim(-.5,2.1)
    clear_output(wait=True)
    plt.plot([xc-6,xc,xc+6],[f2(xc)-df2(xc)*6,f2(xc),f2(xc)+df2(xc)*6],c="k",alpha=.5)
    plt.plot([xc,xcn],[f2(xc),f2(xcn)],ls="--",c="r",alpha=.5,lw=2)
    display(fig)
    time.sleep(.8)
    plt.close()
plt.close()
```

### Método de bisección: una alternativa sin derivadas

Veamos ahora un método que no usa derivadas para encontrar la solución de una ecuación. Este método se conoce como el método de la bisección y consiste en "acorralar" a la solución. Lo primero que hay que hacer si buscamos la solución de una ecuación

$$
f(x_*) = a,
$$

es reescribir la ecuación de manera que lo que busquemos sea una raiz:

$$g(x) = f(x)-a \Rightarrow g(x_*) = 0.$$

El método de la bisección se limita a funciones monótonas en el intervalo en el que buscamos la raiz $[a,b]$. Dado que $x_*$ es raiz de $g(x)$ y $g(x)$ es una función monótona en $[a,b]$, entonces las regiones $x>x_*$ y $x<x_*$ tienen signos distintos.  
El método consiste en ir reduciendo el intervalo original $[a,b]$ hasta que encontremos la raiz. Para ello se siguen los siguientes pasos:

- Comenzamos con un intervalo $\mathcal{I}_0 = [a,b]$ y un punto medio de este intervalo, $c=(a+b)/2$
- Nos fijamos si $g(c)$ tiene el mismo signo que $g(a)$ o que $g(b)$. En el primer caso, actualizamos el intervalo
a $\mathcal{I}_1 = [c,b]$ y en el segundo a $\mathcal{I}_1 = [a,c]$. De manera efectiva, esto reduce el intervalo a la mitad y garantiza que la raiz sigue dentro del intervalo.
- Repetimos lo anterior algún número de veces o mejor aún, hasta alcanzar cierta tolerancia. Por ejemplo que el intervalo final tenga cierta longitud o que $|f(a)-f(b)|$ para el intervalo final sea menor que algún valor.
- La raiz que proponemos será el punto medio del último intervalo.

```python
def biseccion(f,a,b,tol):
    II = [(a,b)]
    c=(a+b)/2
    cc =[c]
    while abs(g1(c)) > tol:
        if np.sign(g1(c)) == np.sign(g1(a)):
            a = c
        else:
            b = c
        c = (a+b)/2
        II.append ((a,b))
        cc.append(c)
    return II,cc
```
